```
# -*- coding: utf-8 -*-
#=========================================
# Recebe e armazena os dados de entrada
#=========================================
enunciado = dict()
#Lê a primeira linha de entrada convertendo para inteiro
enunciado['livros_total'], enunciado['bibliotecas_total'], enunciado['prazo_total'] = map(int, input().split())
#Lê a Segunda linha de entrada convertendo para uma lista de inteiros
enunciado['livros_pontuacao'] = list(map(int, input().split()))

bibliotecas = dict()
biblioteca = dict()
for i in range(enunciado['bibliotecas_total']):    
  #Lê a Terceira linha de entrada convertendo para lista
  apoio = list(map(int,input().split()))

  biblioteca['id'] = i
  biblioteca['qtd_livros'] = apoio[0]
  biblioteca['dias_cadastro'] = apoio[1]
  biblioteca['livros_por_dia'] = apoio[2]  
  biblioteca['ranking'] = 0
  biblioteca['pontuacao_livros'] = list(enunciado['livros_pontuacao'])

  #Lê a Quarta linha de entrada
  biblioteca['livros'] = list(map(int, input().split()))  

  #Excluir da lista de pontos os livros que não tem
  #Ordenar os pontos inversamente
  #Ordenar os livros de acordo com os pontos
  lista_pontos = biblioteca['pontuacao_livros']
  lista_livros = biblioteca['livros']
  apoio2 = list()
  for j in lista_livros:
    apoio2.append(lista_pontos[j])
  biblioteca['pontuacao_livros'] = apoio2

  #Adiciona o dicionario biblioteca no dicionario bibliotecas
  bibliotecas[i] = biblioteca.copy()

#=========================================
# Calcula o ranking das bibliotecas
#=========================================
for biblioteca in bibliotecas.values():
  #Coleta a pontuação dos livros
  lista_pontos = biblioteca['pontuacao_livros']
  soma = 0
  #soma a pontuação
  for pontos in lista_pontos:
    soma += int(pontos)
  #divide pela quantidade de livros
  media = soma / biblioteca['qtd_livros']
  #multiplica pela quantidade de emprestimos e pelos dias de cadastro menos o prazo total
  nota = media * biblioteca['livros_por_dia'] * (enunciado['prazo_total'] - biblioteca['dias_cadastro'])
  biblioteca['ranking'] = nota  


#=================================================
# Ordena a lista de livros e pontos inversamente
#=================================================
dicionario_livros_pontos = dict()
for biblioteca in bibliotecas.values():
  dicionario_livros_pontos = dict(zip(biblioteca['pontuacao_livros'], biblioteca['livros']))

  dic_liv_pts = dict()
  dic_liv_pts = dict(sorted(dicionario_livros_pontos.items(), reverse=True))
  
  biblioteca['livros'] = list(dic_liv_pts.values())
  biblioteca['pontuacao_livros'] = list(dic_liv_pts.keys())
  
#==========================================
# Prepara o arquivo de saída
#=========================================
#Cria a lista de bibliotecas na ordem de serem lidas
ranking_bibliotecas = dict()
for biblioteca in bibliotecas.values():  
  ranking_bibliotecas[biblioteca['id']] = biblioteca['ranking']

apoio_lista = sorted(ranking_bibliotecas.values(), reverse=True)
# function to return key for any value 
def get_key(val): 
    for key, value in ranking_bibliotecas.items(): 
         if val == value: 
             return key 
apoio_lista2 = list()
for vl in apoio_lista:
  apoio_lista2.append(get_key(vl))

lista_saida_bibliotecas = apoio_lista2
#Cria lista de livros na ordem de serem lidos
dic_livros_final = dict()
for x in lista_saida_bibliotecas:
  #Calcula quantos livros cabem na lista considerando o prazo final - prazo da bb
  tamanho__lista = (enunciado['prazo_total'] - bibliotecas[x]['dias_cadastro']) * bibliotecas[x]['livros_por_dia']
  #Cria uma lista só com os livros que serão lidos antes do prazo final
  lista = list(bibliotecas[x]['livros'])
  lista = lista[0:tamanho__lista + 1]
  #Insere as listas em um dicionário
  dic_livros_final[x] = lista.copy()

#Itera pelas listas e exclui os livros que já foram lidos nas listas anteriores
for a in range(len(lista_saida_bibliotecas)-1, 0, -1):
  for b in range(a-1, -1, -1):
    for c in range(len(dic_livros_final[a])-1, -1, -1):      
      if dic_livros_final[a][c] in dic_livros_final[b]:
        del(dic_livros_final[a][c])


#Imprimir resultado final linha 1
x = 0
for val in dic_livros_final:  
  if len(dic_livros_final[val]) != 0:
    x += 1
print(x)

#Imprimir resultado final linha 2
for val in dic_livros_final:  
  final = ''
  if len(dic_livros_final[val]) != 0:
    print(val, len(dic_livros_final[val]))
    for n in range(len(dic_livros_final[val])):
      final += str(dic_livros_final[val][n]) + ' '
    print(final)
```
